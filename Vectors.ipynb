```python
import numpy as np

v = np.array([1, 2, 3, 4])
x = np.array([5, 6, 7, 8])

np.dot(v, x)
```

### 공분산 Covariance

한 개의 변수 값이 변화할 때 다른 변수가 어떤 연관을 가지고 변화하는지 측정하는 것입니다. 

image.png

<img src = 'https://www.simplilearn.com/ice9/free_resources_article_thumb/covx-y.jpg'>


공분산이 음인 경우 변수 X,Y는 음의 상관관계를, 양인 경우에는 양의 상관관계를 가진다고 볼 수 있습니다.

```python
import pandas as pd

df = pd.DataFrame([[170,60], [156,49], [165,52]], columns = ('키', '몸무게'))
df
```

```python
df.cov() #양의 상관관계를 가진다고 말할 수 있습니다.
```

### 상관계수 (Correlation Co-efficient)
위의 공분산은 그 데이터의 스케일에 따라 값이 달라지기 때문에, 비교하기 다소 어렵습니다.
상관계수는 그 범위를 -1 부터 1사이로 조정해줍니다

\begin{align}
cor(X,Y) = r = \frac{cov(X,Y)}{\sigma_{X}\sigma_{Y}} 
\end{align}

```python
df.corr()
```
